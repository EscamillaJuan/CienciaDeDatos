## Ejercicio 1:  Datos de la estación meteorologica de Progreso, Yucatán

```python
df= pd.read_csv("datos/Estacion_PROGRESO_90_dias_19Nov.csv")
df.head()
```

<div class="alert alert-success">

1. Considerando todos los datos hacer una grafica de dispersión para la Temperatura de Aire (en el eje Y) y la Humadad Relativa (en el eje X) 
</ul>
</div>

```python
x_min = df["HumedadRelativa(%)"].min()
x_max = df["HumedadRelativa(%)"].max()
y_min = df["TemperaturaAire(C)"].min()
y_max = df["TemperaturaAire(C)"].max()

fig, ax = plt.subplots(1, 1, figsize=(10, 4), dpi=120)
ax.plot(df["HumedadRelativa(%)"], df["TemperaturaAire(C)"], 'g.')
ax.set_title('Temperatura vs Humedad')
ax.set_xlabel('Humedad relativa (%)')
ax.set_ylabel('Temperatura del aire (°C)')

ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=120)

plt.plot(df["HumedadRelativa(%)"], df["TemperaturaAire(C)"], 'g.')
plt.title('Temperatura vs Humedad')  
plt.xlabel('Humedad relativa (%)')
plt.ylabel('Temperatura del aire (°C)')
plt.xlim = (x_min, x_max)
plt.ylim = (y_min, y_max)
plt.show()
```

# Ejercicio 2: Incendios en México

```python
df= pd.read_csv("datos/Serie_historica_anual_incendios_2017.csv")
print(df.columns)
df.head()
```

<div class="alert alert-success">
    
1. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para los 5 estados más afectados 
    
</div> 

```python
top_5 = df.groupby("Estado")["Total Hectareas"].sum().reset_index().sort_values("Total Hectareas", ascending=False).head()
top_5
plt.figure(figsize=(10, 4), dpi=120)
plt.bar(top_5['Estado'], top_5['Total Hectareas'], color='skyblue')
plt.title('Total de Hectáreas Afectadas por Estado')
plt.xlabel('Estado')
plt.ylabel('Hectáreas Afectadas')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4), dpi=120)
ax.bar(top_5['Estado'], top_5['Total Hectareas'], color='skyblue')
ax.set(title='Rapidez del viento', xlabel='Estado', ylabel='Hectareas Afectadas')
ax.set_xticks(range(len(top_5['Estado'])))
ax.set_xticklabels(top_5['Estado'], rotation=45, ha='right')
plt.show()
```

<div class="alert alert-success">
    
2. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para Yucatán para los meses: abril, mayo, junio, julio y agosto 
    
</div> 

```python
meses_seleccionados = ['abril', 'mayo', 'junio', 'julio', 'agosto']
df_yuc = df[(df['Estado'] == 'Yucatán') & (df['Mes Acontecimiento'].isin(meses_seleccionados))]

plt.figure(figsize=(10, 4), dpi=120)
plt.bar(df_yuc['Mes Acontecimiento'], df_yuc['Total Hectareas'], color='skyblue')
plt.title('Hectáreas afectadas en Yucatán por meses')
plt.xlabel('Mes')
plt.ylabel('Total de Hectáreas Afectadas')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4), dpi=120)
ax.bar(df_yuc['Mes Acontecimiento'], df_yuc['Total Hectareas'], color='skyblue')
ax.set(title='Hectáreas afectadas en Yucatán por meses', xlabel='Mes', ylabel='Total de Hectareas Afectadas')
plt.show()
```

<div class="alert alert-success">
    
3. Considerando todos los datos, hacer un histograma de la duración de los incendios 
    
</div> 

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4), dpi=120)
ax.hist(df['Duracion H'], bins=50, color='blue', alpha=0.4)
ax.set(title='Duracion de los Incendios', xlabel='Duracion H', ylabel='Frecuencia', xlim=(0,30), ylim=(0, 5000))

plt.show()
```

```python
plt.figure(figsize=(10, 4), dpi=120)
plt.hist(df['Duracion H'], bins=50, color='blue', alpha=0.4)
plt.title('Duracion de los Incendios')
plt.xlabel('Duracion H')
plt.ylabel('Frecuencia')
plt.gca().set_xlim(0, 30)
plt.gca().set_ylim(0, 5000)

plt.show()
```

<div class="alert alert-success">
    
4. Considerando todos los datos, hacer una grafica de dispersión entre la duración (eje X) y el total de hectáreas (eje Y)
    
</div> 

```python
x_min = df["Duracion H"].min()
x_max = df["Duracion H"].max()
y_min = df["Total Hectareas"].min()
y_max = df["Total Hectareas"].max()

fig, ax = plt.subplots(1, 1, figsize=(15, 10), dpi=120)
ax.plot(df['Duracion H'], df['Total Hectareas'], 'go')
ax.set(title='Duracion de los Incendios', ylabel='Total de Hectareas', xlabel='Duracion H', xlim=(x_min, x_max), ylim=(y_min,y_max))

plt.figure(figsize=(15,10), dpi=120)
plt.plot(df['Duracion H'], df['Total Hectareas'], 'go')
plt.title('Duracion de los Incendios')  
plt.xlabel('Duracion H')
plt.ylabel('Total de Hectareas')
plt.xlim = (x_min, x_max)
plt.ylim = (y_min, y_max)

plt.show()
```

# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

```python
df_covid= pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201129.csv")
df_covid = df_covid.loc[:,'nombre':]
df_covid = df_covid.set_index('nombre').T
df_covid=df_covid.reset_index()
df_covid = df_covid.rename(columns = {'index': 'Fecha'})
df_covid['Fecha'] = pd.to_datetime(df_covid['Fecha'], format= '%d-%m-%Y')
df_covid = df_covid.set_index('Fecha')
df_covid.head()
```

<div class="alert alert-success">
    
1. Hacer una grafica de barras comparando todos los meses, considerando el total nacional (por mes) 
    
</div> 

```python
df_covid_na = df_covid.resample('M').sum()
df_covid_na['Fecha_str'] = df_covid_na.index.strftime('%m')
plt.figure(figsize=(15, 7), dpi=120)
plt.bar(df_covid_na['Fecha_str'], df_covid_na['Nacional'])
plt.title('Total Nacional por Mes')
plt.xlabel('Mes')
plt.ylabel('Total Nacional')

fig, ax = plt.subplots(1, 1, figsize=(15, 7), dpi=120)
ax.bar(df_covid_na['Fecha_str'], df_covid_na['Nacional'])
ax.set(title='Total Nacional por Mes', xlabel='Mes', ylabel='Total Nacional')
plt.show()
```

<div class="alert alert-success">
    
2. Hacer una grafica de barras comparando  los meses de agosto, septiembre, octubre y noviembre, considerando solo las entidades de Yucatán, Distrito Federal  y Jalisco
    
</div> 

```python
estados = ['YUCATAN', 'DISTRITO FEDERAL', 'JALISCO']
df_nuevo = df_covid[estados].resample('M').sum()
meses = ['2020-08-31', '2020-09-30', '2020-10-31', '2020-11-30']
df_nuevo = df_nuevo.loc[meses]

plt.figure(figsize=(10, 6))
df_nuevo.plot(kind='bar', colormap='viridis', edgecolor='black')
plt.title('Casos en agosto, septiembre, octubre y noviembre')
plt.xlabel('Mes')
plt.ylabel('Casos')
plt.legend()
plt.xticks(rotation=45)

fig, ax = plt.subplots(figsize=(10, 6))
df_nuevo.plot(kind='bar', ax=ax, colormap='viridis', edgecolor='black')
ax.set_title('Casos en agosto, septiembre, octubre y noviembre')
ax.set_xlabel('Mes')
ax.set_ylabel('Casos')
ax.set_xticklabels(meses, rotation=45)
plt.legend()
plt.show()
```
